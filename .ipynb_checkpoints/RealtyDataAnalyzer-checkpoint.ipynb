```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
import seaborn as sns
import nltk
nltk.download()
```

```python
house_data = pd.read_csv("HouseDataFinal.csv", index_col=0, header=0)
house_data
```

```python
data = house_data["Price"] < 1000000
data = house_data[data]

pl = sns.distplot(data["Price"], kde=False, rug=True)
pl
```

```python
pr = sns.boxplot(x=data["Price"])
pr
```

```python
pp = sns.jointplot(x = "Baths", y = "Price", data=house_data.loc[:, ['Price', 'Baths']])
pp
```
