```python
def change_nans(df):
    attr_types = df.dtypes
    for column in df.columns:
        if return_none_type(column, attr_types) is None:
            continue
        for row,col_val in zip(df[column].index, df[column].values):
            try:
                if np.isnan(col_val):
                    df.at[row, column] = return_none_type(column, attr_types)
            except:
                pass
            
    return df
```

```python
def initial_clean_data(data_filename):
    """
    Params:
        data_filename: name of file containing data
    Function:
        Parses file for individual dfs and stores values according to columns, then joins in one master df
    """
    
    columns = []
    houses = []
    with open(data_filename) as file:
        for attrs in reader(file):
            #Check if the line being read contains column definitions
            if attrs[0] == '':
                columns = attrs[1:]
                continue
            else:
                house_dict = {}
                for attr, attr_name in zip(attrs[1:], columns):
                    house_dict[attr_name] = attr
                houses.append(house_dict)

    houses_df = pd.DataFrame.from_dict(houses, orient='columns')
    houses_df.set_index('Address', inplace=True)
    houses_df.drop_duplicates(inplace=True)
    return houses_df
```

```python
def send_csv():
    cleaned_df = initial_clean_data("HouseData.csv")
    cleaned_df.to_csv('HouseDataCleaned.csv')
```

```python
def valid_rows(df, column):
    to_drop = df[[column]].dropna()
    to_drop = to_drop[to_drop[column] != '']
    return to_drop
```

```python
def drop_sparse_columns(df, clean_count=50):
    for column in df:
        valid = valid_rows(df, column)
        if len(valid) < clean_count:
            df.drop(columns=[column], inplace=True)
        else:
            pass
    return df
```

```python
def return_none_type(attr_name, attr_types):
    if attr_types[attr_name] == np.dtype('float64'):
        return 0
    else:
        return None
```
