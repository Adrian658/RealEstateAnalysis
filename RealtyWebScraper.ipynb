```python
def crawl_realtor_com(start_page):
    base_search_url = "https://www.realtor.com/realestateandhomes-search"
    base_detail_url = "https://www.realtor.com/realestateandhomes-detail"
    austin_houses_url = base_search_url + "/Austin_TX"
    
    property_types = ["estate", "plan", "land"]
    
    bad_urls = ["https://www.realtor.com/realestateandhomes-detail/realestateandhomes-detail/Bronte_Smithfield_1617-Frate-Barker-Rd_Austin_TX_78748_P414964638571",
               "https://www.realtor.com/realestateandhomes-detail/realestateandhomes-detail/Guadalupe_Vistas-of-Austin_10216-Laurie-Ln_Austin_TX_78747_P414191714617",
               "https://www.realtor.com/realestateandhomes-detail/realestateandhomes-detail/Tenor_Vistas-of-Austin_10216-Laurie-Ln_Austin_TX_78747_P413111471337",
               "https://www.realtor.com/realestateandhomes-detail/realestateandhomes-detail/Harrison_Sweetwater_6120-Llano-Stage-Trl_Austin_TX_78738_P413998381183",
               "https://www.realtor.com/realestateandhomes-detail/realestateandhomes-detail/Virginia_Easton-Park_7704-Skytree-Dr_Austin_TX_78744_P415547633594"]

    soup = create_soup(austin_houses_url)
    
    if soup is None:
        return None
    
    for i in range(start_page, int(find_page_count(soup))+1):
        house_data = []
        page_url = austin_houses_url + "/pg-" + str(i)
        house_urls_uniq = crawl_index_page(create_soup(page_url))
        for house in house_urls_uniq:
            house_url = base_detail_url + house
            house_soup = create_soup(house_url)
            
            if house_url in bad_urls:
                continue
            
            print("Crawling page: ", house_url)
            for property_type in property_types:
                try:
                    estate_data = crawl_page_estate(house_soup, property_type)
                    estate_data["Property Type"] = property_type
                    break
                except AttributeError as e:
                    pass
            
            estate_data["url"] = house_url
            house_data.append(estate_data)
            estate_data = None
            
        house_data = pd.DataFrame.from_dict(house_data, orient='columns')
        #house_data.set_index("Address", inplace=True)
        
        set_cols=True
        
        house_data.to_csv("HouseData.csv", mode='a', header=set_cols)
            
        print("Finished page: ", i)
            
    return house_data
```

```python
def create_soup(url, parser='html.parser'):
    
    #This header is used to trick the webpage into believing the request is sent from a browser
    headers = ({'User-Agent':
            'Mozilla/5.0 (Macintosh; Intel Mac OS X x.y; rv:42.0) AppleWebKit/537.36 (KHTML, like Gecko) Safari/537.36'}
              )
    req = requests.get(url, headers=headers)
    
    return BeautifulSoup(req.text, parser)
```

```python
def crawl_index_page(souped_page):
    """
    Params:
        souped_page: the soup object containing html of webpage
    """
    
    urls = []
    
    for listing in souped_page.find_all(class_="component_property-card"):
        data_wrap = listing.find(class_="data-wrap")
        if data_wrap != None:
            url = data_wrap.get("data-url")
            urls.append(url)
        
    return urls
```

```python
def crawl_page_estate(souped_page, page_type):
    """
    Params:
        souped_page: the soup object containing html of webpage
    """
    
    estate_data = {}
    
    #Get and store agent information
    estate_data.update(branding_agent_info(souped_page, page_type))
    
    #Get and store main metadata
    estate_data.update(main_metadata(souped_page, page_type))
    
    #Find and store listing section details. Should contain the following:
        #Status, year built, property type, price per sq ft, time listed for, garage
    estate_data.update(main_details(souped_page, page_type))
    
    #Find and store property description if available
    estate_data.update(property_description(souped_page, page_type))

    #Find and store each property detail
    estate_data.update(property_details(souped_page, page_type))
    
    return estate_data
```

```python
def branding_agent_info(souped_page, page_type):
    
    estate_data = {}
    
    #Get and store agent information
    if page_type == 'plan':
        branding_builder_name = remove_junk_chars(souped_page.find(attrs={"data-label": "branding-builder-name"}).text)
        estate_data["Branding Builder Name"] = branding_builder_name
        estate_data["Branding Agent Name"] = None
        estate_data["Branding Office Name"] = None
    else:
        try:
            branding_agent_name = souped_page.find(attrs={"data-label": "branding-agent-name"}).text
            estate_data["Branding Agent Name"] = branding_agent_name
            branding_office_name = souped_page.find(attrs={"data-label": "branding-office-name"}).text
        except AttributeError:
            branding_office_name = None
        estate_data["Branding Office Name"] = branding_office_name
        estate_data["Branding Builder Name"] = None
    
    return estate_data
```

```python
def main_metadata(souped_page, page_type):
    
    estate_data = {}
    
    #Get main metadata
    main_info = souped_page.find(class_="header-main-info")
    price = remove_junk_chars(main_info.find(attrs={"itemprop": "price"}).text, "currency")
    category = main_info.find(attrs={"itemprop": "category"})["content"]
    address = remove_junk_chars(main_info.find(class_="ldp-header-address").text, "general")
    latitude = main_info.find(attrs={"itemprop": "latitude"})["content"]
    longitude = main_info.find(attrs={"itemprop": "longitude"})["content"]
    if page_type != 'land':
        beds = main_info.find(attrs={"data-label": "property-meta-beds"}).find(class_="data-value").text
        baths = remove_junk_chars(main_info.find(attrs={"data-label": "property-meta-bath"}).find(class_="data-value").text, 'number')
        sqft = remove_junk_chars(main_info.find(attrs={"data-label": "property-meta-sqft"}).find(class_="data-value").text, "number")
        estate_data["Beds"] = int(beds)
        estate_data["Baths"] = float(baths)
        estate_data["Sqft"] = float(sqft)
    if page_type != 'plan':
        try:
            lotsize = remove_junk_chars(main_info.find(attrs={"data-label": "property-meta-lotsize"}).find(class_="data-value").text, "number")
            estate_data["Lotsize"] = float(lotsize)
        except AttributeError:
            estate_data["Lotsize"] = None
    
    #Store main metadata
    try:
        estate_data["Price"] = int(price)
        estate_data["Category"] = category
        estate_data["Address"] = address
        estate_data["Latitude"] = float(latitude)
        estate_data["Longitude"] = float(longitude)
    except:
        pass
    
    return estate_data
```

```python
def main_details(souped_page, page_type):
    
    estate_data = {}
    
    #Find and store listing section details. Should contain the following:
        #Status, year built, property type, price per sq ft, time listed for, garage
    listing_detail_section = souped_page.find(class_="listing-section-details")
    listing_details = listing_detail_section.find_all(class_="ldp-key-fact-item")
    for detail in enumerate(listing_details):
        index = detail[0]
        detail_segments = list(detail[1].children)
        detail_segments = [segment for segment in detail_segments if segment != '\n']
        detail_reformatted = reformat_detail(detail_segments[1].text, remove_junk_chars(detail_segments[2].text, "general"))
        
        estate_data[detail_reformatted[0]] = detail_reformatted[1]
        
    return estate_data
```

```python
def property_description(souped_page, page_type):
    
    estate_data = {}
    
    #Find and store property description if available
    property_description_begin = souped_page.find(class_="details-load-more")
    description = property_description_begin.find(id="ldp-detail-romance")
    if description is not None:
        estate_data["description"] = description.text
        
    return estate_data
```

```python
def property_details(souped_page, page_type):
    
    estate_data = {}
    
    listing_detail_section = souped_page.find(class_="listing-section-details")
    
    #Find and clean property details section
    root_removable = "feature-fullscreen-gallery-"
    removable = ["exterior_and_lot_features", "kitchen_and_dining", "bathrooms", "bedrooms"]
    for toremove in removable:
        try:
            listing_detail_section.find(id=root_removable + toremove).decompose()
        except:
            pass
    
    #Variables to store property features
    property_details = clean_text(listing_detail_section.text)
    property_features = []
    
    #Read section and store text that follows 'Property Features', as preceding info is not useful
    start_parsing = False
    for line in iter(property_details.splitlines()):
        if line == "Property Features":
            start_parsing = True
            continue
        if start_parsing:
            property_features.append(line)
        else:
            pass
        
    #Read and store each property detail
    for line in property_features:
        breakdown = line.split(":")
        if len(breakdown) == 1:
            continue
        try:
            breakdown[1] = float(remove_junk_chars(breakdown[1], "currency"))
        except:
            pass
        estate_data[breakdown[0]] = breakdown[1]
        
    return estate_data
```

```python
def clean_text(text):
    """
    Purpose:
        Removes empty lines from input string
    Params:
        text: string to edit
    """
    
    new_text = ""
    for line in iter(text.splitlines()):
        if not line:
            pass
        else:
            new_text = new_text + line + "\n"
    return new_text
```

```python
def reformat_detail(detail_identifier, detail_data):
    """
    Purpose:
        Parses a detail type passed in to modify it's format correspondingly
        Ex. Pass in $5,000 and return 5000
    Params:
        detail: the detail data to modify
        detail_index: index of detail serving as the identifier
        detail_type: type of detail (key or value)
    """
    
    if detail_identifier == "Property Type":
        return detail.replace(" ", "")
    elif detail_identifier == "Price/Sq Ft":
        try:
            return [detail_identifier, int(remove_junk_chars(detail_data, 'currency'))]
        except:
            return [detail_identifier, 0]
    elif detail_identifier == "On realtor.com":
        return ["Hours Listed For", convert_english_time_to_hours(detail_data)]
    elif detail_identifier == "Built":
        return [detail_identifier, int(detail_data)]
        
    return [detail_identifier, detail_data]
```

```python
def convert_english_time_to_hours(time):
    """
    Purpose:
        Convert a relative time expressed in English to hours integer
    Params:
        time: english time string to convert
    """
    
    segments = time.split()
    switcher = {
        "hour": 1,
        "day": 24,
        "week": 24*7,
        "month": 24*7*28,
        "year": 24*7*28*365
    }
    for measure in switcher:
        if measure in segments[1]:
            return int(segments[0]) * switcher[measure]
```

```python
def remove_junk_chars(data, data_type="general"):
    """
    Purpose:
        Remove junk characters from a passed in string
    Params:
        data: string to remove characters from
        data_type: the type of data passed in which will determine which characters are 'junk'
    """
    
    switcher = {
        "currency": [" ", ",", "\n", "$"],
        "general": ["  ", "\n"],
        "number": [",", "+"]
    }
    
    replace_chars = switcher[data_type]
    for char in replace_chars:
        data = data.replace(char, "")
    return data
```

```python
def find_page_count(soup):
    counter = 0
    pagination = soup.find(class_="pagination")
    for child in pagination.children:
        if counter == 9:
            return int(child.text.replace("\n", ""))
        counter+=1
```
